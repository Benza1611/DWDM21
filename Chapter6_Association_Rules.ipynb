##Frequence Itemsets to Association Rule




###เตรียม data สำหรับ (Fequence Pattern) Asdosiation Rule

```python
transactions = data_clean.groupby('InvoiceNo')['StockCode'].apply(list)
```

##Support
Support คือ จำนวน transactions ทั้งหมดที่เราสนใจอยู่ (% ใน itemsets)

```python
from apyori import apriori # หลังจากที่ install package เรียบร้อยแล้วจะสามารถเรียกใช้งานได้
```

```python
frequentItemsets = list(apriori(transactions,min_support = 0.1)) # อย่างน้อย 10 % ของ record ทั้งหมด
frequentItemsets
```

```python
frequentItemsets = list(apriori(transactions,min_support = 0.035)) # ยิ่งลด support จะได้ Rule ออกมาเยอะมากขึ้น
frequentItemsets 
```

```python
frequentItemsets[-1] # มีคนซื้อพร้อมๆกัน
```

#(Quiz7) หา k-itemset ที่มีความน่าสนใจ (โดยพิจารณาลูกค้าเป็นรายคน) พร้อมทั้งอธิบายว่าน่าสนใจยังไง ส่งก่อน 11.25

```python
transactions_1 = data_clean.groupby('CustomerID')['StockCode'].apply(list)
transactions_1
```
